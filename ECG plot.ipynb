```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ecg = pd.read_csv("ecg.csv")
sal = pd.read_csv("saliency.csv")
```

```python
ecg1 = ecg[0:1].T
sal1 = sal[0:1]
```

```python
sal2 = pd.DataFrame()
for i in range(int(ecg1.max() + 20), int(ecg1.min() - 20), -1):
    sal2 = sal2.append(sal1)
    
sal2.index = list(range(int(ecg1.max() + 20), int(ecg1.min() - 20), -1))
```

```python
plt.figure(figsize=(20,3))
plt.plot(ecg1-int(ecg1.min() - 10))
sns.heatmap(sal2, vmin = 0, vmax = 14, yticklabels=False, xticklabels=False)
plt.gca().invert_yaxis()
plt.show()
```
